#  A3 - Comitê de Classificadores: Wine Quality
Pré-processamento e Modelos Individuais

## 1. Importação das Bibliotecas

```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)
```

## 2. Carregamento do Dataset

```python
df = pd.read_csv('WineQT.csv')
print(f'Linhas: {df.shape[0]} | Colunas: {df.shape[1]}')
```

```python
from IPython.display import HTML

def tabela_scroll(df, altura=300):
    estilo = f'overflow-y: scroll; height: {altura}px; display: block;'
    return HTML(f'<div style="{estilo}">{df.to_html()}</div>')

tabela_scroll(df)
```

## 3. Separando Features (X) e Target (Y)

```python
x = df.drop(['quality', 'Id'], axis=1)
y = df['quality']

print(f'Features (X): {x.shape[1]} colunas')
print(f'Target (y): {y.nunique()} classes diferentes')
print(f'Classes: {sorted(y.unique())}')
```

## 4. Normalização dos Dados

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

## 5. Divisão Treino / Teste

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f'Treino: {X_train.shape[0]} amostras')
print(f'Teste:  {X_test.shape[0]} amostras')
```

## 6. Modelo 1 – KNN (K-Nearest Neighbors)

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)
pred_knn = knn.predict(X_test)

print(f'Acurácia do KNN: {accuracy_score(y_test, pred_knn):.2%}')
print()
print(classification_report(y_test, pred_knn, zero_division=0))
```

```python
ConfusionMatrixDisplay(
    confusion_matrix(y_test, pred_knn, labels=sorted(y.unique())),
    display_labels=sorted(y.unique())
).plot(cmap='Blues')
plt.title('Matriz de Confusão – KNN')
plt.tight_layout()
plt.show()
```

## 7. Modelo 2 – Árvore de Decisão

```python
tree = DecisionTreeClassifier(random_state=42, max_depth=10)
tree.fit(X_train, y_train)
pred_tree = tree.predict(X_test)

print(f'Acurácia da Árvore de Decisão: {accuracy_score(y_test, pred_tree):.2%}')
print()
print(classification_report(y_test, pred_tree, zero_division=0))
```

```python
ConfusionMatrixDisplay(
    confusion_matrix(y_test, pred_tree, labels=sorted(y.unique())),
    display_labels=sorted(y.unique())
).plot(cmap='Blues')
plt.title('Matriz de Confusão – Árvore de Decisão')
plt.tight_layout()
plt.show()
```

## 8. Modelo 3 – Naive Bayes

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
pred_nb = nb.predict(X_test)

print(f'Acurácia do Naive Bayes: {accuracy_score(y_test, pred_nb):.2%}')
print()
print(classification_report(y_test, pred_nb, zero_division=0))
```

```python
ConfusionMatrixDisplay(
    confusion_matrix(y_test, pred_nb, labels=sorted(y.unique())),
    display_labels=sorted(y.unique())
).plot(cmap='Blues')
plt.title('Matriz de Confusão – Naive Bayes')
plt.tight_layout()
plt.show()
```

## 9. Modelo 4 – SVM (Support Vector Machine)

```python
svm = SVC(kernel='rbf', C=10, random_state=42)
svm.fit(X_train, y_train)
pred_svm = svm.predict(X_test)

print(f'Acurácia do SVM: {accuracy_score(y_test, pred_svm):.2%}')
print()
print(classification_report(y_test, pred_svm, zero_division=0))
```

```python
ConfusionMatrixDisplay(
    confusion_matrix(y_test, pred_svm, labels=sorted(y.unique())),
    display_labels=sorted(y.unique())
).plot(cmap='Purples')
plt.title('Matriz de Confusão – SVM')
plt.tight_layout()
plt.show()
```

## 10. Modelo 5 – Rede Neural (MLPClassifier)

```python
mlp = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)
mlp.fit(X_train, y_train)
pred_mlp = mlp.predict(X_test)

print(f'Acurácia da Rede Neural: {accuracy_score(y_test, pred_mlp):.2%}')
print()
print(classification_report(y_test, pred_mlp, zero_division=0))
```

```python
ConfusionMatrixDisplay(
    confusion_matrix(y_test, pred_mlp, labels=sorted(y.unique())),
    display_labels=sorted(y.unique())
).plot(cmap='Greens')
plt.title('Matriz de Confusão – Rede Neural')
plt.tight_layout()
plt.show()
```

## 11. Acurácia por Nota de Qualidade

```python
y_test_reset = y_test.reset_index(drop=True)
pred_knn_s   = pd.Series(pred_knn)
pred_tree_s  = pd.Series(pred_tree)
pred_nb_s    = pd.Series(pred_nb)
pred_svm_s   = pd.Series(pred_svm)
pred_mlp_s   = pd.Series(pred_mlp)

print('Acurácia por Nota de Qualidade:')
print('------------------------------------------')
for nota in sorted(y_test_reset.unique()):
    mask = (y_test_reset == nota).values
    if mask.sum() > 0:
        acc_knn  = accuracy_score(y_test_reset[mask], pred_knn_s[mask])
        acc_tree = accuracy_score(y_test_reset[mask], pred_tree_s[mask])
        acc_nb   = accuracy_score(y_test_reset[mask], pred_nb_s[mask])
        acc_svm  = accuracy_score(y_test_reset[mask], pred_svm_s[mask])
        acc_mlp  = accuracy_score(y_test_reset[mask], pred_mlp_s[mask])
        print(f'Nota {nota} ({mask.sum()} amostras):')
        print(f'  KNN:         {acc_knn:.2%}')
        print(f'  Árvore:      {acc_tree:.2%}')
        print(f'  Naive Bayes: {acc_nb:.2%}')
        print(f'  SVM:         {acc_svm:.2%}')
        print(f'  Rede Neural: {acc_mlp:.2%}')
```

## 12. Comparação Geral dos Modelos

```python
resultados = pd.DataFrame({
    'Modelo': ['KNN', 'Árvore de Decisão', 'Naive Bayes', 'SVM', 'Rede Neural'],
    'Acurácia (%)': [
        round(accuracy_score(y_test, pred_knn)  * 100, 2),
        round(accuracy_score(y_test, pred_tree) * 100, 2),
        round(accuracy_score(y_test, pred_nb)   * 100, 2),
        round(accuracy_score(y_test, pred_svm)  * 100, 2),
        round(accuracy_score(y_test, pred_mlp)  * 100, 2),
    ]
}).sort_values('Acurácia (%)', ascending=False)

print(resultados.to_string(index=False))
```

```python
plt.figure(figsize=(9, 5))
sns.barplot(
    data=resultados,
    x='Modelo',
    y='Acurácia (%)',
    hue='Modelo',
    palette='Blues_r',
    legend=False
)
plt.title('Comparação de Acurácia dos Modelos')
plt.ylabel('Acurácia (%)')
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

## 13. Tabela de Previsões

```python
tabela_previsoes = pd.DataFrame(X_test, columns=x.columns)
tabela_previsoes['nota_real']   = y_test.values
tabela_previsoes['knn']         = pred_knn
tabela_previsoes['arvore']      = pred_tree
tabela_previsoes['naive_bayes'] = pred_nb
tabela_previsoes['svm']         = pred_svm
tabela_previsoes['rede_neural'] = pred_mlp

tabela_scroll(tabela_previsoes)
```

## 14. Distribuição das Notas no Conjunto de Teste

```python
plt.figure(figsize=(8, 5))
sns.countplot(x=y_test, hue=y_test, palette='viridis', legend=False)
plt.title('Distribuição das Notas no Conjunto de Teste')
plt.xlabel('Qualidade do Vinho')
plt.ylabel('Número de Amostras')
plt.tight_layout()
plt.show()
```

## 15. Salvando os modelos para o comitê

```python
with open('modelos_treinados.pkl', 'wb') as f:
    pickle.dump({
        'knn':           knn,
        'arvore':        tree,
        'nb':            nb,
        'svm':           svm,
        'rede_neural':   mlp,
        'X_test_scaled': X_test,
        'y_test':        y_test,
        'scaler':        scaler
    }, f)
```

## 16. Comitê por Votação Majoritária

```python
from scipy import stats

votos = np.array([pred_knn, pred_tree, pred_nb, pred_svm, pred_mlp])
pred_comite, _ = stats.mode(votos, axis=0)
pred_comite = pred_comite.flatten()

print(f'Acurácia do Comitê (Votação Majoritária): {accuracy_score(y_test, pred_comite):.2%}')
print()
print(classification_report(y_test, pred_comite, zero_division=0))
```

```python
ConfusionMatrixDisplay(
    confusion_matrix(y_test, pred_comite, labels=sorted(y.unique())),
    display_labels=sorted(y.unique())
).plot(cmap='Oranges')
plt.title('Matriz de Confusão – Comitê (Votação Majoritária)')
plt.tight_layout()
plt.show()
```

## 17. Comitê por Votação Ponderada

```python
acc_knn  = accuracy_score(y_test, pred_knn)
acc_tree = accuracy_score(y_test, pred_tree)
acc_nb   = accuracy_score(y_test, pred_nb)
acc_svm  = accuracy_score(y_test, pred_svm)
acc_mlp  = accuracy_score(y_test, pred_mlp)
pesos = np.array([acc_knn, acc_tree, acc_nb, acc_svm, acc_mlp])
pesos = pesos / pesos.sum()

print('Pesos utilizados:')
for nome, p in zip(['KNN', 'Árvore', 'Naive Bayes', 'SVM', 'Rede Neural'], pesos):
    print(f'  {nome}: {p:.4f}')
print()

classes = sorted(y.unique())
n_amostras = len(y_test)
votos_pond = np.zeros((n_amostras, len(classes)))

for i, pred in enumerate([pred_knn, pred_tree, pred_nb, pred_svm, pred_mlp]):
    for j, amostra in enumerate(pred):
        idx_classe = classes.index(amostra)
        votos_pond[j, idx_classe] += pesos[i]

pred_comite_pond = np.array([classes[np.argmax(linha)] for linha in votos_pond])

print(f'Acurácia do Comitê (Votação Ponderada): {accuracy_score(y_test, pred_comite_pond):.2%}')
print()
print(classification_report(y_test, pred_comite_pond, zero_division=0))
```

```python
ConfusionMatrixDisplay(
    confusion_matrix(y_test, pred_comite_pond, labels=sorted(y.unique())),
    display_labels=sorted(y.unique())
).plot(cmap='Reds')
plt.title('Matriz de Confusão – Comitê (Votação Ponderada)')
plt.tight_layout()
plt.show()
```

## 18. Comparação Final

```python
resultados_final = pd.DataFrame({
    'Modelo': [
        'KNN', 'Árvore de Decisão', 'Naive Bayes', 'SVM', 'Rede Neural',
        'Comitê (Majoritário)', 'Comitê (Ponderado)'
    ],
    'Acurácia (%)': [
        round(acc_knn  * 100, 2),
        round(acc_tree * 100, 2),
        round(acc_nb   * 100, 2),
        round(acc_svm  * 100, 2),
        round(acc_mlp  * 100, 2),
        round(accuracy_score(y_test, pred_comite)      * 100, 2),
        round(accuracy_score(y_test, pred_comite_pond) * 100, 2),
    ]
}).sort_values('Acurácia (%)', ascending=False)

print(resultados_final.to_string(index=False))
```

```python
cores = ['#4C72B0'] * 5 + ['#DD8452', '#C44E52']
plt.figure(figsize=(11, 6))
bars = plt.barh(
    resultados_final['Modelo'],
    resultados_final['Acurácia (%)'],
    color=cores[::-1]
)
plt.xlabel('Acurácia (%)')
plt.title('Comparação Final: Modelos Individuais vs. Comitê')
plt.xlim(0, 100)
for bar in bars:
    w = bar.get_width()
    plt.text(w + 0.5, bar.get_y() + bar.get_height() / 2,
             f'{w:.2f}%', va='center', fontsize=9)
plt.tight_layout()
plt.show()
```
